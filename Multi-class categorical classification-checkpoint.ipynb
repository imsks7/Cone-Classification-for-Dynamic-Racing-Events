Checking distribution of different Class 

Here each class does not have same number of entries. So the model will be biased to the class having highest number of samples. So, we can perform two different task one is undersampling and another one is downsampling to make all the class having equal number of entries so that the model wont be inclined to any specific class. Here we will perform downsampling and all class will have same number of entries which will be equal to the number of entries of the class having minimum entries. 

```python
# Get the count of each category in the 'label_column' column
category_counts = efficientnet_combined['label'].value_counts()

# Print the category counts
print(category_counts)
```

Downsampling the dataset to make distribution equal of all 4 classes 

```python
from sklearn.utils import resample


# Separate the data by class
class_1 = efficientnet_combined[efficientnet_combined['label'] == 1]
class_2 = efficientnet_combined[efficientnet_combined['label'] == 2]
class_3 = efficientnet_combined[efficientnet_combined['label'] == 3]
class_4 = efficientnet_combined[efficientnet_combined['label'] == 4]

# Resample each class to have the same number of entries as the smallest class
smallest_class_size = min(len(class_1), len(class_2), len(class_3), len(class_4))
class_1_resampled = resample(class_1, replace=True, n_samples=smallest_class_size)
class_2_resampled = resample(class_2, replace=True, n_samples=smallest_class_size)
class_3_resampled = resample(class_3, replace=True, n_samples=smallest_class_size)
class_4_resampled = resample(class_4, replace=True, n_samples=smallest_class_size)

# Concatenate the resampled classes into a single DataFrame
resampled_df = pd.concat([class_1_resampled, class_2_resampled, class_3_resampled, class_4_resampled])

# Print the resulting DataFrame
print(resampled_df)
```

```python
# Get the count of each category in the 'label_column' column
category_counts = resampled_df['label'].value_counts()

# Print the category counts
print(category_counts)
```

Model Building
Random Forest

```python
#Seperating Independent and Dependent Variable
X=resampled_df.drop(['label'],axis=1)
y=resampled_df['label']
```

```python
#Performing train test split 80:20
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#Training the model (Random Forest) with default parameters (by default 100 trees will be considered)
from sklearn.ensemble import RandomForestClassifier

classifier_rf = RandomForestClassifier(random_state=0) #setting random state 0 to make reproducible result over different run
classifier_rf.fit(x_train, y_train)
```

```python
#Checking the performance of the model

from sklearn.metrics import accuracy_score,confusion_matrix
#Making prediction on test set
y_pred = classifier_rf.predict(x_test)
#checking the accuracy on test set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

#checking the confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", confusion_mat)
```

Dimensionality reduction using PCA and using the latent component as feature in SVM model 

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_classification

# Compute the PCA components and the explained variance ratio
pca = PCA().fit(X)
explained_variance_ratio = pca.explained_variance_ratio_

# Plot the cumulative explained variance ratio
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
plt.plot(cumulative_variance_ratio)
plt.title("Components vs Variance plot with PCA for efficientNet Dataset")
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance ratio')
plt.show()
```
